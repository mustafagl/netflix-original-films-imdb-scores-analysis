```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

```python
data = pd.read_csv('/kaggle/input/netflix-original-films-imdb-scores/NetflixOriginals.csv', encoding='ISO-8859-1')

data.info()
```

# Veri setine göre uzun soluklu filmler hangi dilde oluşturulmuştur? Görselleştirme yapınız.

```python
long_run_languages=data.groupby("Language", as_index=False)[["Runtime"]].mean().sort_values(by="Runtime",ascending=False).reset_index(drop=True)

display(long_run_languages.head())
```

**En uzun soluklu 4 dilin Pie Chart Gösterimi**

```python
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d})".format(pct, absolute)

explode = (0.1, 0, 0, 0)
plt.title('En uzun soluklu 4 dilin Pie Chart Grafiği')
plt.pie(long_run_languages.head(4).values[:,1],labels=long_run_languages.head(4).values[:,0], explode=explode,autopct=lambda pct: func(pct, long_run_languages.head(4).values[:,1]))
plt.show()
```

**En kısa soluklu 4 dilin Pie Chart gösterimi**

```python
explode = (0.1, 0, 0, 0)
plt.title('En kısa soluklu 4 dilin Pie Chart Grafiği')
plt.pie(long_run_languages.tail(4).values[:,1],labels=long_run_languages.tail(4).values[:,0], explode=explode,autopct=lambda pct: func(pct, long_run_languages.tail(4).values[:,1]))
plt.show()
```

**Tüm dillerin Bar Chart gösterimi**

```python
fig = plt.figure(figsize = (20, 10))

plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large'  
)
plt.bar(long_run_languages.values[:,0], long_run_languages.values[:,1], color ='maroon',
        width = 0.8)
 
plt.xlabel("Diller")
plt.ylabel("Ortalama Uzun Soluk")
plt.title("Uzun Soluklu Diller Bar Chart Grafiği")
plt.show()
```

# 2019 Ocak ile 2020 Haziran tarihleri arasında 'Documentary' türünde çekilmiş filmlerin IMDB değerlerini bulup görselleştiriniz.


```python
data["Premiere_numeric"] = pd.to_datetime(data.Premiere)
documentery_imdb_2019_2020 = data.loc[(data["Genre"] == "Documentary") & (data["Premiere_numeric"] > "2019-01-31") & (data["Premiere_numeric"] < "2020-06-01")].sort_values(by="Premiere_numeric").reset_index(drop=True)

display(documentery_imdb_2019_2020.head(5))
display(documentery_imdb_2019_2020.info())
display(documentery_imdb_2019_2020["IMDB Score"].describe())
```

**Scatter plot ile IMDB Score dağılımı gösterimi**

```python

fig = plt.figure(figsize = (20, 10))

plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large'  
)
plt.yticks(np.arange(0, 10, 0.5),fontsize='x-large')
plt.xlabel("Date",fontsize=20)
plt.ylabel("IMDB Score",fontsize=20)
plt.title("Scatter plot ile IMDB Score dağılımı gösterimi",fontsize=30)
plt.scatter(documentery_imdb_2019_2020["Premiere_numeric"].values, documentery_imdb_2019_2020["IMDB Score"].values)
plt.show()
```

**Histogram ile IMDB Score dağılımı gösterimi**

```python
fig = plt.figure(figsize = (20, 10))
plt.yticks(np.arange(0, 20, 1.0),fontsize=15)
plt.xticks(np.arange(0, 10.5, 0.5),fontsize=15)

plt.hist(documentery_imdb_2019_2020["IMDB Score"].values,bins=20,edgecolor='black',range=(0, 10))
plt.xlabel("IMDB Score", fontsize=20)
plt.ylabel("Film sayısı", fontsize=20)
plt.title("Histogram ile IMDB Score dağılımı gösterimi",fontsize=30)
plt.show() 
```

**Bar Chart ile IMDB Score gösterimi**

```python
fig = plt.figure(figsize = (20, 10))

plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large'  
)

plt.yticks(np.arange(0,15,0.5),horizontalalignment='right',fontweight='light',fontsize='x-large')

plt.bar(documentery_imdb_2019_2020["Title"].values, documentery_imdb_2019_2020["IMDB Score"].values, color ='maroon',
        width = 0.8)
 
plt.xlabel("Title",fontsize=20)
plt.ylabel("IMDB Score",fontsize=20)
plt.title("Bar Chart ile IMDB Score gösterimi",fontsize=30)
plt.show()
```

# IMDB puanı ile 'Runtime' arasında nasıl bir korelasyon vardır? İnceleyip görselleştiriniz.

```python

fig, ax = plt.subplots(figsize = (20, 10))

ax.set_xticks(np.arange(2),labels=["IMDB Score","Runtime"],fontsize=15)
ax.set_yticks(np.arange(2),labels=["IMDB Score","Runtime"],fontsize=15)
for i in range(2):
    for j in range(2):
        text = ax.text(j, i, round(data[["IMDB Score", "Runtime"]].corr().values[i, j],3),
                       ha="center", va="center", color="green",fontsize=20)

ax.set_title("IMDB Score / Runtime korelasyon",fontsize=30)

im = ax.imshow(data[["IMDB Score", "Runtime"]].corr())

ax.figure.colorbar(im, orientation="vertical")
ax.set_xticks(np.arange(3)-.5, minor=True)
ax.set_yticks(np.arange(3)-.5, minor=True)
ax.grid(which="minor", color="w", linestyle='-', linewidth=3)
ax.tick_params(which="minor", bottom=False, left=False)
```

# IMDB Puanı en yüksek olan ilk 10 'Genre' hangileridir? Görselleştiriniz.

```python
imdb_top10_genre=data.groupby("Genre").max().sort_values(by="IMDB Score",ascending=False)[["IMDB Score"]].reset_index().head(10)
display(imdb_top10_genre)
```

**Line plot ile ilk 10 Genre gösterimi**

```python
fig = plt.figure(figsize = (20, 10))

plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large'  
)

plt.yticks(np.arange(0,10,0.25),horizontalalignment='right',fontweight='light',fontsize='x-large')

plt.plot(imdb_top10_genre["Genre"].values, imdb_top10_genre["IMDB Score"].values, color ='red',ls='dotted')
 
plt.xlabel("Genre",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.title("Line plot ile ilk 10 Genre gösterimi",fontsize=30)
plt.grid()

plt.show()
```

# 'Runtime' değeri en yüksek olan ilk 10 film hangileridir? Görselleştiriniz.

```python
runtime_top10_genre=data.sort_values(by="Runtime",ascending=False)[["Title","Runtime"]].reset_index(drop=True).head(10)
display(runtime_top10_genre)
```

**Runtime değeri en yüksek olan ilk 10 filmin Pie Chart Grafiği**

```python
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d})".format(pct, absolute)

#explode = (0.1, 0, 0, 0)
plt.title('Runtime değeri en yüksek olan ilk 10 filmin Pie Chart Grafiği')
plt.pie(runtime_top10_genre["Runtime"].values,labels=runtime_top10_genre["Title"].values, autopct=lambda pct: func(pct,runtime_top10_genre["Runtime"].values))
plt.show()
```

# Hangi yılda en fazla film yayımlanmıştır? Görselleştiriniz.


```python
data['year'] = data['Premiere_numeric'].dt.year
film_count_year = data["year"].value_counts(sort=True).reset_index(drop=False)
film_count_year.columns = ["year","count"]
display(film_count_year)
print("{} yılında en fazla film yayınlanmıştır.".format(film_count_year["year"][0]))
```

**Yıllara göre yayınlanan film sayısının pie chart grafiği**

```python
plt.figure(figsize=(20,10))

plt.title("Yıllara göre yayınlanan film sayısının pie chart grafiği")
plt.pie(film_count_year["count"].values,labels=film_count_year["year"].values, autopct='%.2f')

plt.legend()
plt.tight_layout()

plt.show()
```

# Hangi dilde yayımlanan filmler en düşük ortalama IMBD puanına sahiptir? Görselleştiriniz.
